<h1>Cleaning Of The Data Set</h1>
from this point of th code cleaning the dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('2019.csv', encoding='latin1')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
date_cols = ['AsOfDate', 'ApprovalDate', 'FirstDisbursementDate']
for col in date_cols:
      df[col] = pd.to_datetime(df[col], errors='coerce')

print(date_cols)
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['InitialInterestRate'] = pd.to_numeric(df['InitialInterestRate'], errors='coerce')
df['TermInMonths'] = pd.to_numeric(df['TermInMonths'], errors='coerce')
df['GrossApproval'] = pd.to_numeric(df['GrossApproval'], errors='coerce')
df['SBAGuaranteedApproval'] = pd.to_numeric(df['SBAGuaranteedApproval'], errors='coerce')
```

```python
df.fillna({'FranchiseName': 'Not a Franchise', 'LoanStatus': 'Unknown'}, inplace=True)
```

```python
df['LoanStatus'] = df['LoanStatus'].str.lower().str.strip()
```

```python
df.drop(columns=['BankFDICNumber', 'BankNCUANumber', 'CongressionalDistrict'], inplace=True, errors='ignore')
```

```python
df.to_csv("cleaned_loan_data.csv", index=False)
print("Data cleaning done. Saved as 'cleaned_loan_data.csv'.")
```

```python
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot to visualize outliers
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[['GrossApproval', 'InitialInterestRate', 'TermInMonths']])
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.countplot(x=df["LoanStatus"])
plt.title("Class Distribution")
plt.show()
```
